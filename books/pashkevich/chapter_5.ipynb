## Задачи для самостоятельного решения

### Задача 1

Выборка представлена следующими значениями X: 10, 2, 6, 6, 4, 2

Вычислите несмещенную  выборочную оценку дисперсии и стандартного отклонения

**Решение:**

1) Вычислим выборочное среднее:

$$\overline{x} = \frac{10 + 2 + 6 + 6 + 4 + 2}{6} = \frac{30}{6} = 5$$

2) Вычислим несмещенную оценку дисперсии:

$$s_{несмещ}^2 = \frac{1}{n - 1}\sum\limits_{i=1}^{n}(x_i - \overline{x})$$

$$s_{несмещ}^2 = \frac{(10 - 5)^2 + (2 - 5)^2 + (6 - 5)^2 + (6 - 5)^2 + (4 - 5)^2 + (2 - 5)^2}{5} = \frac{46}{5} = 9.2$$

3) Вычислим несмещенное стандартное отклонение:

$$s_{несмещ} = \sqrt{9,2} \approx 3.03$$

### Задача 2. Спорт

В ходе небольшого опроса, проведённого на одном из занятий по математической статистики (в 2013 г.) среди студентов факультета прикладной политологии НИУ "Высшая школа экономики", участникам семинара было предложено ответить на несколько коротких вопросов. Одним из таких вопросов был следующий: "Укажите, пожалуйста, сколько примерно часов в неделю Вы тратите на спорт и различные физические упражнения (включая бег, бассейн, утреннюю зарядку, ролики, танцы и прочие виды активности). Напишите с точностью до получаса". Объём выборки юношей составил 8 человек, девушек - 11 человек. Ответы распределились следующим образом:

    Юноши: 2, 5.5, 6.5, 10, 2.5, 5, 7, 7
    Девушки: 7, 4, 2, 5, 4, 0, 0, 2, 3, 2, 5
    
Для обеих выборок найдите выборочные оценки среднего, дисперсии (как смещённую, так и несмещённую), оцените величину смещения. Рассчитайте несмещённые выборочные оценки стандратного отклонения

**Решение:**

0) X - случайная величина того, сколько юноши тратят на спорт; часы
 
    Y - случайная величина того, сколько девушки тратят на спорт; часы

1) Рассчитаем выборочные оценки среднего:

$$\overline{x} = \frac{2 + 5.5 + 6.5 + 10 + 2.5 + 5 + 7 + 7}{8} = \frac{45.5}{8} \approx 5.69$$

$$\overline{y} = \frac{7 + 4 + 2 + 5 + 4 + 0 + 0 + 2 + 3 + 2 + 5}{11} = \frac{32}{11} \approx 3.09$$

2) Рассчитаем смещённую дисперсию:

$$s^2_{смещ;x} = \frac{1}{n_x}\sum\limits_{i=1}^{n_x}(x_i - \overline{x})^2$$

$$s^2_{смещ;x} = \frac{(2 - 5.69)^2 + (5.5 - 5.69)^2 + (6.5 - 5.69)^2 + (10 - 5.69)^2 + (2.5 - 5.69)^2 + (5 - 5.69)^2 + (7 - 5.69)^2 + (7 - 5.69)^2}{8} \approx 5.87$$

$$s^2_{смещ;y} = \frac{(7 - 3.09)^2 + (4 - 3.09)^2 + (2 - 3.09)^2 + (5 - 3.09)^2 + (4 - 3.09)^2 + (0 - 3.09)^2}{11} + \frac{(0 - 3.09)^2 + (2 - 3.09)^2 + (3 - 3.09)^2 + (2 - 3.09)^2 + (5 - 3.09)^2}{11} \approx 4.26$$

3) Рассчитаем несмещённую дисперсию:

$$s^2_{несмещ;x} = \frac{1}{n_x - 1}\sum\limits_{i=1}^{n_x}(x_i - \overline{x})^2$$

$$s^2_{несмещ;x} = \frac{(2 - 5.69)^2 + (5.5 - 5.69)^2 + (6.5 - 5.69)^2 + (10 - 5.69)^2 + (2.5 - 5.69)^2 + (5 - 5.69)^2 + (7 - 5.69)^2 + (7 - 5.69)^2}{7} \approx 6.7$$

$$s^2_{несмещ;y} = \frac{(7 - 3.09)^2 + (4 - 3.09)^2 + (2 - 3.09)^2 + (5 - 3.09)^2 + (4 - 3.09)^2 + (0 - 3.09)^2}{10} + \frac{(0 - 3.09)^2 + (2 - 3.09)^2 + (3 - 3.09)^2 + (2 - 3.09)^2 + (5 - 3.09)^2}{10} \approx 4.69$$

4) Рассчием величину смещения:

    X: $\frac{8}{7} \approx 1.1428 \rightarrow 14.28 \%$

    Y: $\frac{11}{10} = 1.1 \rightarrow 10 \%$
    
5) Рассчитаем несмещённые выборочные оценки стандартного отклонения:

$$s_{несмещ;x} = \sqrt{6.7} \approx 2.59$$

$$s_{несмещ;y} = \sqrt{4.69} \approx 2.16$$
