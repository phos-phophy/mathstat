### Задача 1. Мечты

Студентами факультета социологии на одном из семинаров по математической статистике был инициирован блиц-опрос на семейную тему. Один из опросов, на которые требовалось дать ответ, звучал так: "О каком количестве детей в своей будущей семье Вы мечтаете? Укажите, пожайлуста, числом". Ответы молодёжи были сведены в таблицу:

| X, число детей | 0 | 1 | 2 | 3 | 4 | 5 | Всего |
| --- | --- | --- | --- | --- | --- | --- | --- |
| Число ответивших | 6 | 8 | 45 | 21 | 5 | 3 | 88 |

Вычислите выборочное среднее. Охарактеризуйте разброс значений с помощью разных мер вариации, таких как выборочное стандартное отклонение, коэффициент вариации. Визуализируйте имеющийся вариационный ряд, построив полигон распределения и кумулятивную кривую.

**Решение:**

1) Выборочное среднее: 
$$\overline{x} = \frac{0 * 6 + 1 * 8 + 2 * 45 + 3 * 21 + 4 * 5 + 5 * 3}{88} = \frac{196}{88} = 2$$

2) Стандартное отклонение: 
$$\sigma^2 = \frac{1}{n}\sum\limits_{i=1}^n{(x_i -\overline{x})}^2$$

$$\sigma^2 = \frac{6 * (0 - 2)^2 + 8 * (1 - 2) ^ 2 + 45 * (2 - 2) ^ 2 + 21 * (3 - 2) ^ 2 + 5 * (4 - 2) ^ 2 + 3 * (5 - 2) ^ 2}{88} = \frac{100}{88} \approx 1.13$$

$$\sigma \approx 1.06$$

3) Коэффициент вариации:
$$V_x = \frac{\sigma}{\overline{x}} * 100 \% = \frac{1.06}{2} * 100 \% = 53 \%$$

### Задача 2. Студенты и автомобиль

Пусть имеются следующие данные о том, какова доля студентов университета, приезжающих на занятия на личном автомобиле:

| Факультет | А | Б | В | Г | Д | Е | Ж |
| --- | --- | --- | --- | --- | --- | --- | --- |
| Доля студентов в общей численности факультета | 0.12 | 0.06 | 0.17 | 0.09 | 0.10 | 0.08 | 0.08 |

Определите, чему равна, в среднем, на каждом факультете доля студентов, приезжающих на своём автомобиле. Рассчитайте выборочное стандартное отклонение и выборочный коэффициент вариации. О чём свидетельствует полученные значение коэффициента вариации? Что можно сказать о доле автомобилистов среди студентов разных факультетов?

**Решение:**

1) Выборочное среднее:
$$\overline{x} = \frac{0.12 + 0.06 + 0.017 + 0.09 + 0.10 + 0.08 + 0.08}{7} = \frac{0.70}{7} = 0.10$$

2) Стандартное отклонение:
$$\sigma^2 = \frac{1}{n}\sum\limits_{i=1}^{n}(x - \overline{x})^2$$

$$\sigma^2 = \frac{(0.12 - 0.10) ^ 2 + (0.06 - 0.10) ^ 2 + (0.17 - 0.10) ^ 2 + (0.09 - 0.10) ^ 2 + (0.10 - 0.10) ^ 2 + 2 * (0.8 - 0.10) ^ 2}{7} = \frac{0.0078}{7} \approx 0.0011$$

$$\sigma \approx 0.033$$

3) Коэффициент вариации:

$$V_x = \frac{\sigma}{\overline{x}} * 100 \% = \frac{0.033}{0.10} * 100 \% = 33 \%$$

Данное значение коэффициента вариации говорит о том, что данные обладают **высокой однородностью**, т.е. доля автомобилистов среди студентов сильно не различается от факультета к факультету

| 40.0 | 47.5 | 52.2 | 53.8 | 55.2 | 55.5 | 56.8 |

### Задача 4. Стоимость жилья

Имеются следующие данные о стоимости вторичного жилья, предлагаемого к продаже на территории Южного федерального округа РФ (Данные Федеральной службы государственной статистики по Волгоградской области (http://www.volgastat.ru/federal/ufb14/default.aspx)). В таблице приведены средние цены 1 м$^2$ общей площади жилых помещений типовых квартир на конец I квартала 2012 г.

| Регион России | Средняя цена 1 м$^2$, тыс. руб. |
|---|---|
| Республика Адыгея | 24216 |
| Республика Калмыкия | 33432 |
| Краснодарский край | 47466 |
| Астраханская обл. | 29852 |
| Волгоградская обл. | 31508 |
| Ростовская обл. | 47758 |

Найдите среднюю цену 1 м$^2$ вторичного жилья в регионах Южного федерального округа. Рассчитайте выборочное стандартное отклонение и выборочный коэффициент вариации. О чём свидетельствует полученное значение коэффициента вариации? Можно ли считать, что стоимость вторичного жилья в ЮФО однородная (похожая)? Или нет?

**Решение:**

1) Выборочное среднее:

$$\overline{x} = \frac{24216 + 33432 + 47466 + 29852 + 31508 + 47758}{6} \approx 35705.33$$

2) Стандартное отклонение:

$$\sigma^2 = \frac{1}{n}\sum\limits_{i=1}^{n}(x_i - \overline{x})^2$$

$$\sigma^2 = \frac{(24216 - 35705.33)^2 + (33432 - 35705.33)^2 + (47466 - 35705.33)^2 + (29852 - 35705.33)^2}{6} + \frac{(31508 - 35705.33)^2 + (47758 - 35705.33)^2}{6} \approx 78771999.55$$

$$\sigma \approx 8875.36$$

3) Коэффициент вариации:

$$V_x = \frac{\sigma}{\overline{x}} * 100 \% = \frac{8875.36}{35705.33} * 100 \% \approx 24.85 \%$$

Можно считать, что данные являются однородными
