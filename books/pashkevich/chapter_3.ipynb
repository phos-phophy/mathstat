### Задача 2

Вычислите коэффициент корреляции Пирсона между двумя признаками X и Y:

|  |  |  |  |  |  |  |
| --- | --- | --- | --- | --- | --- | --- |
| X | 2 | 6 | 7 | 5 | 9 | 7 |
| Y |  9 | 3 | 1 | 5 | 0 | 0 |

**Решение:**

1) Вычислим выборочные средние:

$$\overline{x} = \frac{2 + 6 + 7 + 5 + 9 + 7}{6} = 6$$

$$\overline{y} = \frac{9 + 3 + 1 + 5 + 0 + 0}{6} = 3$$

2) Вычислим ковариацию:

$$cov(X, Y) = \frac{1}{n}\sum\limits_{i=1}^{n}(x_i - \overline{x})(y_i - \overline{y})$$

$$cov(X, Y) = \frac{(2 - 6)*(9 - 3) + (6 - 6)*(3 - 3) + (7 - 6)*(1 - 3) + (5 - 6)*(5 - 3) + (9 - 6)*(0 - 3) + (7 - 6)*(0 - 3)}{6} = -6.(6)$$

3) Вычислим дисперсии:

$$\mathbb{D}_x = \frac{1}{n}\sum\limits_{i=1}^{n}(x_i - \overline{x})^2$$

$$\mathbb{D}_x = \frac{(2 - 6)^2 + (6 - 6)^2 + (7 - 6)^2 + (5 - 6)^2 + (9 - 6)^2 + (7 - 6)^2}{6} = 4.(6)$$

$$\mathbb{D}_y = \frac{(9 - 3)^2 + (3 - 3)^2 + (1 - 3)^2 + (5 - 3)^2 + (0 - 3)^2 + (0 - 3)^2}{6} = 10.(3)$$

4) Вычислим корреляцию:

$$r_{Пирсон} = \frac{cov(X,Y)}{\sqrt{\mathbb{D}_x}\sqrt{\mathbb{D}_y}} = -\frac{6.(6)}{\sqrt{4.(6)}\sqrt{10.(3)}} \approx -0.96$$

### Задача 3. Физика и химия

Десять школьников написали две экзаменационные работы - по физике и химии. Каждая работа включала 20 заданий. Для каждого учащегося зафиксировано число верно выполненных заданий по двум предметам:

| Школьник | Миша | Петя | Оля | Рома | Ира | Нина | Катя | Лиза | Ваня | Коля |
| --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- |
| X: число верно выполенных заданий по физике | 10 | 5 | 3 | 17 | 15 | 12 | 10 | 20 | 14 | 14 |
| Y: число верно выполенных заданий по химии | 18 | 12 | 20 | 10 | 15 | 15 | 20 | 5 | 8 | 7 |

Вычислите коэффициент линейной корреляции Пирсона между числом верно выполенных заданий по физике и числом верно выполненных заданий по химии. В какой мере они коррелируют?

**Решение:**

1) Считаем выборочные средние:

$$\overline{x} = \frac{10 + 5 + 3 + 17 + 15 + 12 + 10 + 20 + 14 + 14}{10} = 12$$

$$\overline{y} = \frac{18 + 12 + 20 + 10 + 15 + 15 + 20 + 5 + 8 + 7}{10} = 13$$

2) Посчитаем ковариацию:

$$cov(X, Y) = \frac{1}{n}\sum\limits_{i=1}^{n}(x_i - \overline{x})(y_i - \overline{y})$$

$$cov(X, Y) = \frac{(10 - 12) * (18 - 13) + (5 - 12) * (12 - 13) + (3 - 12) * (20 - 13) + (17 - 12) * (10 - 13) + (15 - 12) * (15 - 13)}{10} + \frac{(12 - 12) * (15 - 13) + (10 - 12) * (20 - 13) + (20 - 12) * (5 - 13) + (14 - 12) * (8 - 13) + (14 - 12) * (7 - 13)}{10} = -17.5$$

3) Посчитаем дисперсии:

$$\mathbb{D}_x = \frac{1}{n}\sum\limits_{i=1}^{n}(x_i - \overline{x})^2$$

$$\mathbb{D}_x = \frac{(10 - 12)^2 + (5 - 12)^2 + (3 - 12)^2 + (17 - 12)^2 + (15 - 12)^2 + (12 - 12)^2 + (10 - 12)^2 + (20 - 12)^2 + (14 - 12)^2 + (14 - 12)^2}{10} = 24.4$$

$$\mathbb{D}_y = \frac{(18 - 13)^2 + (12 - 13)^2 + (20 - 13)^2 + (10 - 13)^2 + (15 - 13)^2 + (15 - 13)^2 + (20 - 13)^2 + (5 - 13)^2 + (8 - 13)^2 + (7 - 13)^2}{10} = 26.6$$

4) Посчитаем коэффициент корреляции:

$$r_{Пирсон} = \frac{cov(X, Y)}{\sqrt{\mathbb{D}_x}\sqrt{\mathbb{D}_y}}$$

$$r_{Пирсон} = -\frac{17.5}{\sqrt{24.4}\sqrt{26.6}} \approx -0.68$$

Значение коэффициента корреляции ($0.4 \leq 0.68 < 0.7$) говорит нам об умеренно выраженной тесноте линейной связи между X и Y

Можно ли считать две эти социальные характеристики коррелированными? Рассчитайте коэффициент линейной корреляции Пирсона. Найдите коэффициент ранговой корреляции Спирмена.

**Решение:**

0) Положим, что X - число разводов на 1000 браков, а Y - среднедушевой доход, руб

1) Рассчитаем выборочное среднее:

$$\overline{x} = \frac{477 + 493 + 507 + 444 + 428 + 463 + 459 + 523 + 587 + 527 + 574 + 563 + 566 + 549}{14} \approx 511.43 $$

$$\overline{y} = \frac{17677 + 10195 + 11055 + 18158 + 12423 + 10885 + 19422 + 13385 + 16358 + 13398 + 12700 + 20279 + 11961 + 12905}{14} \approx 14342.93$$

2) Рассчитаем дисперсии:

$$\mathbb{D}_x = \frac{1}{n}\sum\limits_{i=1}^{n}(x_i - \overline{x})^2$$

$$\mathbb{D}_x = \frac{(477 - 511.43)^2 + (493 - 511.43)^2 + (507 - 511.43)^2 + (444 - 511.43)^2 + (428 - 511.43)^2 + (463 - 511.43)^2 + (459 - 511.43)^2}{14} + \frac{(523 - 511.43)^2 + (587 - 511.43)^2 + (527 - 511.43)^2 + (574 - 511.43)^2 + (563 - 511.43)^2 + (566 - 511.43)^2 + (549 - 511.43)^2 }{14} \approx 2514.1$$

$$\mathbb{D}_y = \frac{(17677 - 14342.93)^2 + (10195 - 14342.93)^2 + (11055 - 14342.93)^2 + (18158 - 14342.93)^2 + (12423 - 14342.93)^2}{14} + \frac{(10885 - 14342.93)^2 + (19422 - 14342.93)^2 + (13385 - 14342.93)^2 + (16358 - 14342.93)^2 + (13398 - 14342.93)^2}{14} + \frac{(12700 - 14342.93)^2 + (20279 - 14342.93)^2 + (11961 - 14342.93)^2 + (12905 - 14342.93)^2}{14} \approx 10476811.49$$

3) Рассчитаем ковариацию:

$$cov(X, Y) = \frac{1}{n}\sum\limits_{i = 1}^{n}(x_i - \overline{x})(y_i - \overline{y})$$

$$cov(X, Y) \approx -5276.83$$

4) Рассчитаем коэффициент корреляции:

$$r_{Пирсон} = \frac{cov(X, Y)}{\sqrt{\mathbb{D}_x}\sqrt{\mathbb{D}_y}}$$

$$r_{Пирсон} = -\frac{5276.83}{\sqrt{2514.1}\sqrt{10476811.49}} \approx -0.0325$$

Значит, между характеристиками нет линейной зависимости
