### 1. Генеральная совокупность и исследование её свойств

***Генеральная совокупность*** - совокупность всех объектов (единиц), относительно которых предполагается делать выводы при изучении конкретной задачи.

1) *Главной задачей исследования* является **изучение свойств генеральной совокупности**:

    * Математическое ожидание (среднее)
    * Дисперсия (вариация)
    * Стандартное отклонение
    * Медиана
    * Квартильный размах и др.
    
2) Но зачастую **невозможно собрать всю необходимую информацию** обо всех объектах из генеральной совокупности для изучения этих свойств, поэтому мы **не можем действовать напрямую**

### 2. Выборочная совокупность

Тем не менее нам всё-таки доступна некоторая часть объектов из генеральной совокупности, их мы и будем изучать.

***Выборка (выборочная совокупность)*** - часть объектов генеральной совокупности (её подмножество), отобранных по по определённым правилам и подлежащих изучению в пределах программы исследовательского проекта.

Таким образом, мы можем сформулировать решение нашей задачи следующим образом:

### 3. Решение главной задача исследования

    Изучение проявления статистических закономерностей на выборке и перенос (обобщение) результатов с выборки на генеральную совокупность
    
Иными словами:

1) **Изучаем выборку**

2) **Переносим результаты** наблюдений на генеральную совопкупность
