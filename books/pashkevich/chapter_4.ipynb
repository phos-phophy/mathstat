## Задачи для самостоятельного решения

### Задача 1. Антикварный магазин

Средняя сумма чека в некотором антикварном магазине составляет 1070 усл. ед., а стандартное отклонение - 320 усл. ед. Какова вероятность того, что очередной покупатель совершит покупку на сумму от 600 до 900 усл. ед.? (если считать, что сумма покупки -случайная величина, которая описывается нормальным законом распределения с указанными параметрами)

**Решение:**

$X - $ сл. величина суммы покупки, усл. ед

$X \sim \mathcal{N}(1070, 320^2)$

$$\mathbb{P}(600 \leq X \leq 900) - ?$$

$$\mathbb{P}(600 \leq X \leq 900) = \mathbb{P}(\frac{600 - 1070}{320} \leq \frac{X - 1070}{320} \leq \frac{900 - 1070}{320}) = \mathbb{P}(\frac{600 - 1070}{320} \leq Z \leq \frac{900 - 1070}{320}) = $$

$$\mathbb{P}(-1.47 \leq Z \leq -0.53) = \mathbb{P}(0 \leq Z \leq 1.47) - \mathbb{P}(0 \leq Z \leq 0.53) = $$

$$0.4292 - 0.2019 = 0.2273$$

**Ответ:** $\mathbb{P}(600 \leq X \leq 900) = 0.2273$

### Задача 2. Мороженое

В тёплое время года на покупку мороженого семья тратит в среднем 520 руб. в неделю. Стандартное отклонение составляет 107 руб. Какова вероятность того, что в следующе неделю семья потратит на мороженое от 240 до 300 руб.? (если считать, что расходы семьи на мороженое описываются нормальным законом распределения с указанными параметрами)

**Решение:**

$X - $ сл. величина суммы покупки, усл. ед

$X \sim \mathcal{N}(520, 107^2)$

$$\mathbb{P}(240 \leq X \leq 300) - ?$$

$$\mathbb{P(240 \leq X \leq 300)} = \mathbb{P}(\frac{240 - 520}{107} \leq \frac{X - 520}{107} \leq \frac{300 - 520}{107}) = \mathbb{P}(\frac{240 - 520}{107} \leq Z \leq \frac{300 - 520}{107}) = $$

$$\mathbb{P}(-2.61 \leq Z \leq -2.05) = \mathbb{P}(0 \leq Z \leq 2.61) - \mathbb{P}(0 \leq Z \leq 2.05) = $$

$$0.4955 - 0.4798 = 0.0157$$

**Ответ:** $\mathbb{P}(240 \leq X \leq 300) = 0.0157$

### Задача 3. Кофейня

Средняя сумма чека в некоторой кофейне составляет 810 руб, а стандартное отклонение - 230 руб. Какова вероятность, что очередной посетитель кофейни сделает заказ на сумму от 700 до 1100 руб.? (если считать, что сумма чека - случайная величина, которая описывается нормальным законом распределения с указанными параметрами) 

**Решение:**

$X - $ сл. величина суммы покупки, усл. ед

$X \sim \mathcal{N}(810, 230^2)$

$$\mathbb{P}(700 \leq X \leq 1100) - ?$$

$$\mathbb{P}(700 \leq X \leq 1100) = \mathbb{P}(\frac{700 - 810}{230} \leq \frac{X - 810}{230} \leq \frac{1100 - 810}{230}) = \mathbb{P}(\frac{700 - 810}{230} \leq Z \leq \frac{1100 - 810}{230}) = $$

$$\mathbb{P}(-0.48 \leq Z \leq 1.26) = \mathbb{P}(0 \leq Z \leq 1.26) + \mathbb{P}(0 \leq Z \leq 0.48) = $$

$$0.3962 + 0.1844 = 0.5806$$

**Ответ:** $\mathbb{P}(700 \leq X \leq 1100) = 0.5806$
